```python
import openai
import pandas as pd

# OpenAI API 키 설정
openai.api_key = "YOUR_API_KEY"

# 프롬프트 생성 함수
def gen(text, title):
    prompt = (
    f"""
    현재 제목: {title}
    본문: {text}

    당신은 기사 헤드라인 작성자입니다. 당신의 목표는 기사 헤드라인을 독자적으로 분석하고, 선정적인 요소를 제거한 새로운 헤드라인을 작성하는 것입니다.

    주어진 예시를 참고하여 두 가지 과제를 수행하세요:

    선정적인 헤드라인 예시:
    (예시 1) ‘애로부부’ 손민혁, 김세정 닮은꼴 아내 김정민…화끈한 28금 침실 토크 / (선정성 수치) 0.986635745 / (선정적인 이유) "화끈한 28금 침실 토크"는 성적인 뉘앙스를 직접적으로 표현함.
    (예시 2) "성적 의도로 넣었다"...항문에 12cm 유리컵 집어넣은 네팔 유부남 / (선정성 수치) 0.986573756 / (선정적인 이유) "성적 의도"와 "항문에 12cm 유리컵"은 지나치게 선정적이고 불쾌감을 유발할 수 있음.
    (예시 3) 화장실 창문으로 머리 말리는 女 알몸 촬영 시도한 30대의 최후 / (선정성 수치) 0.986496866 / (선정적인 이유) "女 알몸"은 불필요한 선정적 표현을 사용함.
    (예시 4) 남편의 10년 ‘오피스 와이프’ 잡는 결정적 증거…“부정행위 하는 모습 찍어야”(애로부부) 0.986452639 / (선정성 수치) / (선정적인 이유) "부정행위 하는 모습 찍어야"는 불필요하게 사생활을 강조하며 자극적으로 표현함.
    (예시 5) “오늘은 내 애인”…20대 여성 부사관 어깨 만지고 ‘첫 경험’ 물은 상사 / (선정성 수치) 0.986382723 / (선정적인 이유) "첫 경험"과 상사의 부적절한 발언이 성적으로 민감한 주제를 직접 언급함.
    (예시 6) 아이돌 최초 ‘누드집 발간’ 의혹에 “엉덩이 탱글탱글해” 발끈한 전진 / (선정성 수치) 0.986277759 / (선정적인 이유) "엉덩이 탱글탱글해"는 과도하게 성적인 표현을 사용함.
    (예시 7) 태연, 브라톱 입고 과감한 노출 / (선정성 수치) 0.986263394/ (선정적인 이유) "과감한 노출"은 불필요하게 외모와 노출을 강조함.
    (예시 8) 보드카와 먹는 캐비어… 첫 키스와 같은 ‘설렘의 맛’ [김셰프의 씨네퀴진] / (선정성 수치) 0.986192405/ (선정적인 이유) "첫 키스와 같은"은 불필요하게 감각적이고 선정적으로 묘사함.
    (예시 9) 첫 남친이 게이·핑크머리 의사…베일 벗은 '나는 솔로' 17기 / (선정성 수치) 0.986074209 / (선정적인 이유) "첫 남친이 게이"는 편견을 조장할 수 있는 표현이며 불필요한 사생활 강조임.
    (예시 10) 이세영, 아찔한 한뼘 비키니 자태 "예쁜 내 힙과 허벅지" / (선정성 수치) 0.986042798 / (선정적인 이유) "아찔한 한뼘 비키니"와 "예쁜 내 힙과 허벅지"는 과도하게 외모와 신체를 강조함.

    비선정적인 헤드라인 예시:
    (예시 1) ‘애로부부’ 손민혁, 김정민 부부의 솔직한 부부생활 이야기 / (선정성 수치) 0.01067
    (예시 2) 유리컵 부상 사건...네팔 남성의 의학적 이슈 / (선정성 수치) 0.020196
    (예시 3) 화장실에서 여성 사생활 침해 시도...30대 남성 처벌 / (선정성 수치) 0.024006
    (예시 4) 남편의 오랜 친분 관계...부부간 신뢰 회복의 필요성 / (선정성 수치) 0.004656
    (예시 5) 상사의 부적절 발언과 행동...여성 부사관 피해 호소 / (선정성 수치) 0.006762
    (예시 6) 아이돌 사진집 발간 논란…전진의 입장 표명 / (선정성 수치) 0.0054
    (예시 7) 태연, 패션 스타일로 주목받아 / (선정성 수치) 0.004845
    (예시 8) 보드카와 캐비어...독특한 풍미의 조화 [김셰프의 씨네퀴진] / (선정성 수치) 0.094982
    (예시 9) '나는 솔로' 17기...다양한 참가자들의 이야기 / (선정성 수치) 0.006951
    (예시 10) 이세영, 비치웨어 스타일로 화제 / (선정성 수치) 0.036679


    당신의 과제는 다음과 같습니다:

    1. 첫 번째 과제: 주어진 기사 헤드라인이 선정적이라면, 선정적인 이유를 한 줄로 설명하세요. 선정적이지 않다면 "선정적이지 않음"이라고만 답하세요.
    2. 두 번째 과제: 선정적인 이유를 기반으로 헤드라인을 비선정적으로 수정하세요.

    최종적으로 아래 형식에 맞춰 답변을 작성하세요:

    - 선정적인 이유: [선정적 이유 또는 "선정적이지 않음"]
    - 비선정적 제목: [수정된 제목 또는 기존 제목]
    """
    )
    return prompt

# OpenAI API 호출 및 결과 처리 함수
def analyze_and_generate(text, title):
    prompt = gen(text, title)

    try:
        # OpenAI API 호출
        response = openai.ChatCompletion.create(
            model="gpt-4-turbo",
            messages=[
                {"role": "system", "content": "You are a professional news headline evaluator."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.7,
            max_tokens=150
        )

        # 응답 결과에서 선정적인 이유와 새로운 제목 추출
        response_text = response.choices[0].message['content'].strip()
        reason = response_text.split("- 선정적인 이유: ")[1].split("\n")[0].strip()
        new_title = response_text.split("- 비선정적 제목: ")[1].strip()

        return reason, new_title

    except Exception as e:
        print(f"OpenAI API 호출 오류: {e}")
        return "오류 발생", "오류 발생"

# 데이터 처리 및 저장
def process_and_save(input_file, output_file):
    # 엑셀 파일 읽기
    df = pd.read_excel(input_file)

    # 새로운 열 추가를 위한 리스트
    reasons = []
    new_titles = []

    # 각 행에 대해 API 호출 및 처리
    for index, row in df.iterrows():
        title = row['제목']  # 기존 제목
        text = row['본문']  # 기사 본문

        # OpenAI API 호출 및 결과 저장
        reason, new_title = analyze_and_generate(text, title)
        reasons.append(reason)
        new_titles.append(new_title)

    # 데이터프레임에 결과 열 추가
    df['선정성 이유'] = reasons
    df['새로운 제목'] = new_titles

    # 처리된 결과를 새로운 엑셀 파일로 저장
    df.to_excel(output_file, index=False)
    print(f"처리가 완료되었습니다. 파일이 저장되었습니다: {output_file}")

# 사용 예시
input_file = 'KcELECTRA_제목전처리X_수치화_상위100개.xlsx'  # 입력 파일 경로
output_file = "gpt4turbo_전처리X_생성.xlsx"  # 출력 파일 경로

process_and_save(input_file, output_file)
```

```python
import torch.nn.functional as F
import torch
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_save_path = '/content/drive/MyDrive/saved_model/KcELECTRA'

# 1. 모델과 토크나이저 불러오기
model = AutoModelForSequenceClassification.from_pretrained(model_save_path)
tokenizer = AutoTokenizer.from_pretrained(model_save_path)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.eval()
model.to(device)
```

```python
df = pd.read_excel("gpt4turbo생성최종.xlsx")
```

```python
titles = df['새로운 제목'].astype(str).tolist()
```

```python
def predict_probabilities(title):
    # 입력 텍스트를 토큰화
    inputs = tokenizer(title, return_tensors="pt", padding=True, truncation=True, max_length=512).to(device)

    # 모델로 예측 수행
    with torch.no_grad():
        outputs = model(**inputs)
        probabilities = torch.softmax(outputs.logits, dim=1)  # 소프트맥스 적용

    # '1' 클래스(선정성) 확률 반환
    return probabilities[0][1].item()
```

```python
# 제목별 선정성 확률 계산
df['새 선정성 확률'] = df['새로운 제목'].apply(predict_probabilities)

# 결과 확인
print(df[['제목', '새로운 제목', '새 선정성 확률']].head())
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from transformers import BertTokenizer, BertModel
import torch

# BERT 모델 및 토크나이저 로드
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

# 본문과 새로운 제목 임베딩 계산 함수
def get_bert_embedding(text):
    with torch.no_grad():
        inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True)
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1)  # 문장 임베딩
    return embedding

# 본문과 새로운 제목 간의 코사인 유사도 계산 함수
def calculate_similarity(text1, text2):
    embedding1 = get_bert_embedding(text1)
    embedding2 = get_bert_embedding(text2)
    similarity = cosine_similarity(embedding1.numpy(), embedding2.numpy())
    return similarity[0][0]

# 평가 실행 코드
def evaluate_content_similarity(df):
    similarities = []
    for _, row in df.iterrows():
        content = row['본문']  # 본문
        generated_title = row['새로운 제목']  # 생성된 제목

        # 코사인 유사도 계산
        similarity = calculate_similarity(content, generated_title)
        similarities.append(similarity)

    df['본문-새로운제목 유사도'] = similarities  # 유사도 추가
    return df

# 엑셀 파일 로드 및 평가 수행
df = pd.read_excel('gpt4turbo_전처리X_생성.xlsx')
df = evaluate_content_similarity(df)
```
